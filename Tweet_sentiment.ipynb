```python
def remove_special_characters(df1):
  for column in df1.columns:
    df1[column]=df1[column].apply(lambda x:re.sub(r'[^\w\s\d]','', x))
  return df1

df =remove_special_characters(df1)
```

```python
X = df.drop('label',axis=1)
y=df['label']
```

```python
X.shape,y.shape
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
vocab_size=300
```

```python
tweets = X.copy()
```

```python
tweets
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(tweets)):
    review =re.sub('[^a-zA-Z]', ' ', tweets['Tweets'][i])
    review =review.lower()
    review= review.split()

    review= [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review =' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,vocab_size)for words in corpus]
onehot_repr
```

```python
onehot_repr[1]
```

```python
# Embedding
from tensorflow.keras.layers import Embedding

from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[25:32]
```

```python
from tensorflow.keras.models import Sequential

from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
embedding_vector=40 ##features representation
model.add(Embedding(vocab_size,embedding_vector,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
len(embedded_docs),y.shape
```

```python
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.3, random_state=42)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```
